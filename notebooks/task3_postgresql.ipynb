```python
import pandas as pd

from sqlalchemy import create_engine
```

```python
df = pd.read_csv("../data/raw/final_processed_reviews.csv")
```

```python
df.head()
```

```python
banks_df = pd.DataFrame({
    'bank_id': [1, 2, 3],
    'bank_name': [
        'Commercial Bank of Ethiopia',
        'Bank of Abyssinia',
        'Dashen Bank'
    ],
    'app_name': [
        'CBE Mobile',
        'BOA Mobile',
        'Dashen Super App'
    ]
})
```

```python
bank_mapping = {
    'CBE': 1,
    'BOA': 2,
    'Dashen': 3
}

df['bank_id'] = df['bank'].map(bank_mapping)
```

```python
df.head()
```

```python
reviews_df = df[[
    'bank_id',
    'review',
    'rating',
    'date',
    'sentiment_label',
    'sentiment_score',
    'identified_theme',
    'source'
]]
```

```python
reviews_df.columns = [
    'bank_id',
    'review_text',
    'rating',
    'review_date',
    'sentiment_label',
    'sentiment_score',
    'identified_theme',
    'source'
]
```

```python
df.head()
```

```python
engine = create_engine(
    "postgresql://new@localhost:5432/bank_reviews"
)
```

```python
banks_df.to_sql(
    'banks',
    engine,
    if_exists='replace',
    index=False
)
```

```python
reviews_df.to_sql(
    'reviews',
    engine,
    if_exists='replace',
    index=True,
    index_label='review_id'
)
```
